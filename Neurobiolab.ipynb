```python
import pandas as pd
from scipy.stats import skewtest
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#import data set
url = 'https://raw.githubusercontent.com/nellyfadil/Neurobio/main/Final%20Neurobio%20Data%20-%20Sheet1.csv'
alz_data = pd.read_csv(url)
alz_data
```

```python
#Create Variables
Sex= alz_data.iloc[:,2]
Apoe_e4 = alz_data.iloc [:,3]
Diagnosis = alz_data.iloc[:,-2]
MAPT_intensity_HIP = alz_data.iloc[:,-1]

print(Apoe_e4)
```

```python
#Create pie chart for grouping varbiables
#sex
colors = ['Royalblue', 'Red']
alz_data['sex'].value_counts().plot(kind = 'pie', colors = colors)
plt.ylabel('sex')
plt.show()

#diagnosis
plt.show()
colors = ['mistyrose', 'lightcoral','red', 'brown']
alz_data['nincds_arda_diagnosis'].value_counts().plot(kind='pie', colors=colors)
plt.ylabel('DSM IV Diagnosis')

```

```python
df = pd.DataFrame(alz_data)
plt.figure(figsize=(10,6))
boxplot = df.boxplot(column ='MAPT_intensity_HIP', by=['sex'], grid=False, patch_artist=True)

plt.title('Box Plot of MAPT Intensity by Sex')
plt.xlabel('Sex')
plt.ylabel('MAPT Log2 Intensity')

# Show the plot
plt.show()
```

```python
df = pd.DataFrame(alz_data)
plt.figure(figsize=(10,6))
boxplot = df.boxplot(column ='MAPT_intensity_HIP', by=['Apoe_e4'], grid=False, patch_artist=True)

plt.title('Box Plot of MAPT Intensity by Apoe_e4')
plt.xlabel('Apoe Expression')
plt.ylabel('MAPT Log2 Intensity')

# Show the plot
plt.show()
```
