```python
import pandas as pd
from scipy.stats import skewtest
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#import data set
url = 'https://raw.githubusercontent.com/nellyfadil/Neurobio/main/Final%20Neurobio%20Data%20-%20Sheet1.csv'
alz_data = pd.read_csv(url)
alz_data
```

```python
#Create pie chart for grouping varbiables
#sex
colors = ['Royalblue', 'Red']
alz_data['sex'].value_counts().plot(kind = 'pie', colors = colors)
plt.ylabel('sex')
plt.show()

#diagnosis
plt.show()
colors = ['mistyrose', 'lightcoral','red', 'brown']
alz_data['nincds_arda_diagnosis'].value_counts().plot(kind='pie', colors=colors)
plt.ylabel('DSM IV Diagnosis')

```

```python
df = pd.DataFrame(alz_data)
plt.figure(figsize=(10,6))
ax = df.boxplot(column ='MAPT_intensity_HIP', by=['sex'], grid=False, patch_artist=True
                    ,color = dict(whiskers = 'r', medians = 'r')
                    ,boxprops = dict(facecolor = 'white', color = 'blue')
                  )
box_colors = ['lightblue', 'lightgreen']

plt.title('Box Plot of MAPT Intensity by Sex')
plt.xlabel('Sex')
plt.ylabel('MAPT Log2 Intensity')

plt.show()
```

```python
df = pd.DataFrame(alz_data)
plt.figure(figsize=(10,6))
boxplot = df.boxplot(column ='MAPT_intensity_HIP', by=['apo_e4_allele'], grid=False, patch_artist=True
                    ,color = dict(whiskers = 'r', medians = 'r')
                    ,boxprops = dict(facecolor = 'white', color = 'blue')
                  )
box_colors = ['lightblue', 'lightgreen']


plt.title('Box Plot of MAPT Intensity by Apoe')
plt.xlabel('Apoe_4')
plt.ylabel('MAPT Log2 Intensity')

plt.show()
```
