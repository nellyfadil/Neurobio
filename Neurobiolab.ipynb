```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#import data set
url = 'https://raw.githubusercontent.com/nellyfadil/Neurobio/main/Data%20Neurobio.csv'
alz_data = pd.read_csv(url)
alz_data
```

```python
#Create Variables
Sex= alz_data.iloc[:,4]
Dementia = alz_data.iloc[:,18]
MAPT_zscore_forebrain = alz_data.iloc[:,-2]
MAPT_zscore_hippocampus = alz_data.iloc[:,-1]
```

```python
#Create Histograms for distribution of MAPT in the white matter in forebrain & hippocampus
fig, ax = plt.subplots(1,2,figsize=(12,7))

ax[0].hist(MAPT_zscore_forebrain, color = 'darkslategrey')
ax[0].set_xlabel('MAPT_zscore_forebrain')
ax[0].set_ylabel('Counts')
ax[1].hist(MAPT_zscore_hippocampus, color = 'skyblue')
ax[1].set_xlabel('MAPT_zscore_hippocampus')
ax[1].set_ylabel('Counts')
```

```python
#distribution of zscores
values_plus_minus_one_z_score = len(MAPT_zscore_forebrain[MAPT_zscore_forebrain['MAPT_zscore_forebrain'].between(-1, 1)==True])
percent_values_plus_minus_one_z_score = values_plus_minus_one_z_score/len(MAPT_zscore_forebrain)*100
percent_values_plus_minus_one_z_score

MAPT_zscore_forebrain['us_z-score'].hist(color='slategray')
plt.title("Standard Normal Distribution", y=1.015, fontsize=22)
plt.xlabel("z-score", labelpad=14)
plt.ylabel("frequency", labelpad=14);
```

```python
#Create pie chart for grouping varbiables
#sex
colors = ['Royalblue', 'Red']
alz_data['sex'].value_counts().plot(kind = 'pie', colors = colors)
plt.ylabel('sex')
plt.show()

#dementia
plt.show()
colors = ['mistyrose', 'lightcoral','red', 'brown']
alz_data['nincds_arda_diagnosis'].value_counts().plot(kind='pie', colors=colors)
plt.ylabel('DSM Diagnosis')

```
